```python
from pyspark.sql import SparkSession

# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame sederhana
data = [("Ali", 34), ("Budi", 23), ("Citra", 29), ("Dina", 45)]
columns = ["Nama", "Usia"]
df = spark.createDataFrame(data, columns)

# Menampilkan DataFrame
df.show()

# Filtering data
df_filtered = df.filter(df['Usia'] > 30)
df_filtered.show()

# Menghitung rata-rata usia
from pyspark.sql.functions import avg
df.groupBy().agg(avg("Usia")).show()

# Mengurutkan data berdasarkan usia
df_sorted = df.orderBy("Usia", ascending=False)
df_sorted.show()
```

```python
from pyspark.sql import SparkSession
import pandas as pd

spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

data = [("Ali", 34, "Parkir liar", "Tidur", "PS 5"), ("Budi", 23, "Driver", "Silat", "Pria"), ("Citra", 29, "Youtuber", "Makan", "Pria"), ("Dina", 45, "Penjual buah", "Berkebun", "Wanita")]
columns = ["Nama", "Usia", "Pekerjaan", "Hobi", "Gender"]
df = spark.createDataFrame(data, columns)


data_pandas = {
    "Nama": ["Ali", "Budi", "Citra", "Dina"],
    "HP" : ["Sumsuang", "Oddo", "Kingfinix", "Xiaomay"]
}
df_pandas = pd.DataFrame(data_pandas)

# Mengonversi DataFrame dari PySpark ke Pandas
df_pandas_from_spark = df.toPandas()

#gabung
gabung = pd.merge(df_pandas_from_spark, df_pandas, on="Nama")
print("Hasil penggabungan pyspark dan pandas :")
print("\n",gabung)

#rata
rata = gabung["Usia"].mean()
print(f"\nRata-rata usia : {rata} ")
```

```python
from pyspark.sql import SparkSession
import pandas as pd

spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

data = [("Ali", 34, "Parkir liar", "Tidur", "PS 5"), ("Budi", 23, "Driver", "Silat", "Pria"), ("Citra", 29, "Youtuber", "Makan", "Pria"), ("Dina", 45, "Penjual buah", "Berkebun", "Wanita")]
columns = ["Nama", "Usia", "Pekerjaan", "Hobi", "Gender"]
df = spark.createDataFrame(data, columns)


data_pandas = {
    "Nama": ["Ali", "Budi", "Citra", "Dina"],
    "HP" : ["Sumsuang", "Oddo", "Kingfinix", "Xiaomay"]
}
df_pandas = pd.DataFrame(data_pandas)

# Mengonversi DataFrame dari PySpark ke Pandas
df_pandas_from_spark = df.toPandas()

#gabung
gabung = pd.merge(df_pandas_from_spark, df_pandas, on="Nama")
print("Hasil penggabungan pyspark dan pandas :")
print("\n",gabung)

#max
max = gabung["Usia"].max()
print(f"\nUsia tertua : {max} ")
```
